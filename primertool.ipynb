```python
import primertool as pt
import ipywidgets as widgets
from IPython.display import display

primer_order_dataframe = None

def on_value_change(change):
    if change['name'] == 'kuerzel_change':
        pt.config.KUERZEL = change['new']
    elif change['name'] == 'variant_change':
        pt.config.VARIANT = change['new']
    elif change['name'] == 'nm_number_change':
        pt.config.NM_NUMBER = change['new']
    elif change['name'] == 'exon_number_change':
        pt.config.EXON_NUMBER = change['new']
    elif change['name'] == 'chromosome_change':
        pt.config.CHROMOSOME = change['new']
    elif change['name'] == 'start_change':
        pt.config.START = change['new']
    elif change['name'] == 'end_change':
        pt.config.END = change['new']
    elif change['name'] == 'genome_change':
        pt.config.GENOME = change['new']

# KUERZEL
widget_kuerzel = widgets.Text(
    value=pt.config.KUERZEL,
    placeholder='Enter your initials (e.g. XY)',
    description='Kürzel:',
    disabled=False
)
widget_kuerzel.observe(on_value_change, names='kuerzel_change')

# VARIANT
widget_variant = widgets.Text(
    value=pt.config.VARIANT,
    placeholder='Enter variant in HGVS notation (e.g. NM_000410.3:c.845G>A)',
    description='HGVS variant:',
    disabled=False,
    layout = widgets.Layout(width='95%')
)
widget_variant.observe(on_value_change, names='variant_change')

# NM_NUMBER
widget_nm_number = widgets.Text(
    value=pt.config.NM_NUMBER,
    placeholder='Enter NM number (e.g. NM_000451)',
    description='NM number:',
    disabled=False,
    layout = widgets.Layout(width='95%')
)
widget_nm_number.observe(on_value_change, names='nm_number_change')

# EXON_NUMBER
widget_exon_number = widgets.IntText(
    value=pt.config.EXON_NUMBER,
    description='Exon number:',
    disabled=False
)
widget_exon_number.observe(on_value_change, names='exon_number_change')

# CHROMOSOME
widget_chromosome = widgets.Text(
    value=pt.config.CHROMOSOME,
    placeholder='Enter chromosome (e.g. chr12)',
    description='Chromosome:',
    disabled=False
)
widget_chromosome.observe(on_value_change, names='chromosome_change')

# START
widget_start = widgets.IntText(
    value=pt.config.START,
    description='Start:',
    disabled=False
)
widget_start.observe(on_value_change, names='start_change')

# END
widget_end = widgets.IntText(
    value=pt.config.END,
    description='End:',
    disabled=False
)
widget_end.observe(on_value_change, names='end_change')

# GENOME
widget_genome = widgets.ToggleButtons(
    options=['hg19', 'hg38'],
    description='Genome assembly:',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltips=['hg19', 'hg38'],
    value='hg38',
    # icons=['check'] * 2
)
widget_genome.observe(on_value_change, names='genome_change')

out = widgets.Output(layout={'border': '1px solid black'})

def on_button_mutation_clicked(b):
    with out:
        if widget_variant.value and widget_genome.value:
            global primer_order_dataframe
            primer_order_dataframe = pt.PrimerMutation(widget_variant.value, widget_genome.value).create_primer()
            display(primer_order_dataframe.style.hide_index())
        else:
            print('Please enter a variant and select a genome version')
            
def on_button_exon_clicked(b):
    with out:
        if widget_nm_number.value and widget_exon_number.value:
            global primer_order_dataframe
            primer_order_dataframe = pt.PrimerExon(widget_nm_number.value, widget_exon_number.value, widget_genome.value).create_primer()
            display(primer_order_dataframe.style.hide_index())
        else:
            print('Please enter an NM number, an exon number and select a genome version')
        
def on_button_gene_clicked(b):
    if widget_nm_number.value:
        pt.PrimerGen(widget_nm_number.value, widget_genome.value).create_primer()
    else:
        print('Please enter an NM number and select a genome version')

def on_button_genomic_position_clicked(b):
    if widget_chromosome.value and widget_start.value and widget_end.value:
        pt.PrimerGenomicPosition(widget_chromosome.value, widget_start.value, widget_end.value, widget_genome.value).create_primer()
    else:
        print('Please enter a chromosome, a start and an end position and select a genome version')

button_mutation = widgets.Button(
    description='Find primers',
    disabled=False,
    button_style='success',
    tooltip='Find primers for given variant',
    icon='magnifying-glass', # (FontAwesome names without the `fa-` prefix)
)
button_mutation.on_click(on_button_mutation_clicked)

button_exon = widgets.Button(
    description='Find primers',
    disabled=False,
    button_style='success',
    tooltip='Find primers for given exon',
    icon='magnifying-glass' # (FontAwesome names without the `fa-` prefix)
)
button_exon.on_click(on_button_exon_clicked)

button_gene = widgets.Button(
    description='Find primers',
    disabled=False,
    button_style='success',
    tooltip='Find primers for given gene',
    icon='magnifying-glass' # (FontAwesome names without the `fa-` prefix)
)
button_gene.on_click(on_button_gene_clicked)

button_genomic_position = widgets.Button(
    description='Find primers',
    disabled=False,
    button_style='success',
    tooltip='Find primers for given genomic position',
    icon='magnifying-glass' # (FontAwesome names without the `fa-` prefix)
)
button_genomic_position.on_click(on_button_genomic_position_clicked)

title_primer_mutation = widgets.HTML(
    value="<h1>PrimerMutation</h1> Find primers for a given variant.",
    placeholder='PrimerMutation',
)
title_primer_exon = widgets.HTML(
    value="<h1>PrimerExon</h1> Find primers for a given exon.",
    placeholder='PrimerExon',
)
title_primer_gene = widgets.HTML(
    value="<h1>PrimerGene</h1> Find all primers for a given gene.",
    placeholder='PrimerGene',
)
title_primer_genomic_position = widgets.HTML(
    value="<h1>PrimerGenomicPosition</h1> Find primers based on the chromosome, start and end position of the variant.",
    placeholder='PrimerGenomicPosition',
)

# MAIN
tab_titles = ['PrimerMutation', 'PrimerExon', 'PrimerGene', 'PrimerGenomicPosition']
children = [
    widgets.VBox([title_primer_mutation, widget_variant, button_mutation]),  # PrimerMutation
    widgets.VBox([title_primer_exon, widget_nm_number, widget_exon_number, button_exon]),  # PrimerExon
    widgets.VBox([title_primer_gene, widget_nm_number, button_gene]),  # PrimerGen
    widgets.VBox([title_primer_genomic_position, widget_chromosome, widget_start, widget_end, button_genomic_position])  # PrimerGenomicPosition
]
tab_menu = widgets.Tab(children=children, titles=tab_titles, layout = widgets.Layout(width='60%'))
```

```python
display(widget_kuerzel, widget_genome, tab_menu, out)
```
